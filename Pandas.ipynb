# concat/merge

```python
d1={
    'id':['1','2','3','4','5'],
    'Feature1':['A','C','E','G','I'],
    'Feature2':['B','D','F','H','J']}
```

```python
df1=pd.DataFrame(d1,columns=['id','Feature1','Feature2'])
```

```python
df1
```

```python
d1=pd.DataFrame({
    'id':['1','2','3','4','5'],
    'Feature1':['A','C','E','G','I'],
    'Feature2':['B','D','F','H','J']})
```

```python
d1
```

```python
df2=pd.DataFrame({
    'id':['1','2','6','7','8'],
    'Feature1':['K','M','O','Q','S'],
    'Feature2':['L','N','P','R','T']
})
```

```python
df2
```

```python
df3=pd.DataFrame({
    'id':['1','2','3','4','5','7','8','9','10','11'],
    'Feature3':[12,13,14,15,16,17,15,12,13,23]
})
```

```python
df3
```

```python
df1,df2,df3
```

```python
df_row=pd.concat([df1,df2])
df_row
```

```python
df_row_index=pd.concat([df1,df2], ignore_index=True)
df_row_index
```

```python
frames=[df1,df2]
df_keys = pd.concat(frames,keys=['x','y'])
df_keys
```

```python
df_row
```

```python
df3
```

```python
pd.merge(df_row,df3,on="id")
```

```python
df_merge_col = pd.merge(df_row,df3,on='id')
df_merge_col
```

```python
pd.merge(df_row_index,df3,on='id')
```

```python
df_outer = pd.merge(df1,df2, on='id', how='outer')
df_outer
```

```python
df_inner = pd.merge(df1,df2, on='id', how='inner')
df_inner
```

```python
df_suffix = pd.merge(df1,df2, left_on='id',right_on='id',how='outer',
suffixes=('_left','_right'))
df_suffix
```

```python
df_right = pd.merge(df1,df2,on='id',how='right')
df_right
```

```python
df_left = pd.merge(df1,df2, on='id',how='left')
df_left
```
